```python
import requests
from bs4 import BeautifulSoup


url = "http://www.chucknorrisfacts.fr/facts/top/1"
html = requests.get(url)
```

```python
# Make a request on the above web page.

text = html.text
text
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
# What is the response code? What is this error?

html
```

This means that the server has returned HTLM content.

```python
# Bypass the protection by specifying an existing browser. And retry your request.

items_container = soup.find_all("div", {"class", "card"})
items_container
```

```python
# Find the tag to identify a block including a joke. How many joke-blocks are there? (there should be 20).

len(items_container)
```

```python
# Displays the 8th joke-block using the .text attribute.

items_container[7].p.text
```

```python
# Use the .find method with the appropriate tag to isolate only the 8th joke, without the other elements 

items_jokes = soup.find_all(class_= "card-text")
items_jokes[7].text
```

```python
# Do the same to isolate only the rating result of the 8th joke.

items_rating = soup.find_all("span")
items_rating[7].text
```

```python
# Create an empty dictionary, then create a loop that will record in this dictionary each joke as a key and each corresponding rating result as a value.

joke_rating ={}

for i in range(len(items_container)):
  joke_rating[items_container[i].p.text] = items_container[i].span.text
print(joke_rating)
```

```python
# Transform this dictionary into a DataFrame with 2 columns: joke and rating result. It must have 20 rows: one per joke.

from pandas.io.parsers.readers import is_index_col
import pandas as pd

df = pd.DataFrame(joke_rating.items(), columns=['Jokes', 'Rating_Result'])
df
```
